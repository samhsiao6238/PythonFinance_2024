### 顯示

```python
import cv2
import matplotlib.pyplot as plt

# 讀取圖像
img1 = cv2.imread("house.jpg")  # 彩色讀取
img2 = cv2.imread("house.jpg", 0)  # 灰色讀取

# 將 BGR 圖像轉換為 RGB 圖像
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

# 顯示彩色圖像
# 創建一個圖形窗口
plt.figure(figsize=(10, 5))
# 創建子圖，顯示彩色圖像和灰色圖像
plt.subplot(1, 2, 1)
plt.imshow(img1_rgb)
plt.title('MyPicture1')
# 隱藏坐標軸
plt.axis('off')

# 顯示灰色圖像
plt.subplot(1, 2, 2)
plt.imshow(img2, cmap='gray')
plt.title('MyPicture2')
plt.axis('off')

# 顯示圖像
plt.show()
```

### BGR 轉 RGB

```python
import cv2
import matplotlib.pyplot as plt

# BGR 讀取
img = cv2.imread("house.jpg")

# 檢查圖像是否成功讀取
if img is None:
    print("Error: 無法讀取文件 'mountain.jpg'")
else:
    # 顯示 BGR 圖像
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('BGR Color Space')
    plt.axis('off')

    # BGR 轉 RGB
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 顯示 RGB 圖像
    plt.subplot(1, 2, 2)
    plt.imshow(img_rgb)
    plt.title('RGB Color Space')
    plt.axis('off')

    # 顯示圖像
    plt.show()
```

### BGR 到 HSV

```python
import cv2
import matplotlib.pyplot as plt

# BGR 讀取
img = cv2.imread("house.jpg")

# 檢查圖像是否成功讀取
if img is None:
    print("Error: 無法讀取文件 'street.jpg'")
else:
    # 顯示 BGR 圖像
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('BGR Color Space')
    plt.axis('off')

    # BGR 轉 HSV
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # 顯示 HSV 圖像
    plt.subplot(1, 2, 2)
    plt.imshow(img_hsv)
    plt.title('HSV Color Space')
    plt.axis('off')

    # 顯示圖像
    plt.show()
```
