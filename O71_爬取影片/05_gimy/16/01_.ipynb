```python
import subprocess

# 設定目標 m3u8 URL
m3u8_url = "https://v.gsuus.com/play/BeX8YY8e/index.m3u8"

# 設定 ffmpeg 命令，帶有必要的 HTTP 標頭
ffmpeg_cmd = [
    'ffmpeg',
    '-headers', 'User-Agent: Mozilla/5.0\r\nReferer: https://example.com\r\n',
    '-i', m3u8_url,
    '-c', 'copy',
    'downloaded_video.mp4'
]

# 執行下載
try:
    subprocess.run(ffmpeg_cmd, check=True)
    print("下載完成！")
except subprocess.CalledProcessError as e:
    print(f"下載失敗：{e}")
```
