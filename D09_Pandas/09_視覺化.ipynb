# 視覺化

```python
import pandas as pd
import matplotlib.pyplot as plt

# 建立一個 DataFrame
df = pd.DataFrame({
    'Year': [2020, 2021, 2022, 2023, 2024],
    'Sales': [100, 250, 200, 400, 500]
})
```

```python
# 繪製折線圖
plt.plot(df['Year'], df['Sales'])
```

```python
# 繪製折線圖
df.plot(
    x='Year',
    y='Sales',
    kind='line',
    figsize=(8, 7),
    title='Sales Trend'
)
```

# 直方圖

```python
import pandas as pd
import matplotlib.pyplot as plt

# 假設數據
data = {'Scores': [88, 92, 80, 89, 90, 78, 84, 76, 95, 85]}
df = pd.DataFrame(data)

# 繪製直方圖：顯示分數出現頻率
ax = df.plot.hist(
    bins=5,
    title='Histogram',
    xlabel='Scores',
    ylabel='Frequency'
)

# 顯示圖形
plt.show()
```

# 散點圖

```python
import pandas as pd
import matplotlib.pyplot as plt

# 建立數據
data = {
    'Age': [22, 25, 27, 28, 22, 35, 37, 40, 46, 52, 58],
    'Salary': [18000, 23000, 27000, 35000, 22000, 36000, 38000, 48000, 59000, 70000, 85000]
}

# 轉換為 DataFrame
df = pd.DataFrame(data)

# 輸出數據
print('查看數據：')
print(df, '\n')

# 繪製散點圖
ax = df.plot.scatter(x='Age', y='Salary', color='DarkBlue', label='Age vs Salary')

# 設置圖表標題和軸標籤
plt.title('Age vs Salary Scatter Plot')  # 圖表標題
plt.xlabel('Age (Years)')                # X軸標籤
plt.ylabel('Salary (Dollars)')           # Y軸標籤

# 顯示圖形
plt.show()
```
