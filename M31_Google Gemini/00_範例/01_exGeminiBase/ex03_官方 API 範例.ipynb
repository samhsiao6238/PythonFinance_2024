```python
# ç”¨æ–¼è™•ç†æ–‡ä»¶è·¯å¾‘
import pathlib
# å°å…¥ Google Gemini API çš„ SDK
import google.generativeai as genai
import PIL.Image  # ç”¨æ–¼è®€å–åœ–ç‰‡
import os
from dotenv import load_dotenv

load_dotenv()

# è¨­ç½® API é‡‘é‘°ï¼Œè«‹æ›´æ”¹ç‚ºè‡ªå·±çš„å¯¦éš›é‡‘é‘°
GOOGLE_API_KEY = os.getenv("GEMINI_API_KEY")
# é…ç½® Google API ä½¿ç”¨è‡ªå·±çš„é‡‘é‘°
genai.configure(api_key=GOOGLE_API_KEY)

# åˆå§‹åŒ–æ¨¡å‹ï¼Œé€™è£¡ä½¿ç”¨ 'gemini-1.5-flash'
model = genai.GenerativeModel('gemini-1.5-flash')

# è®€å–åœ–ç‰‡æ–‡ä»¶ä¸¦å»ºç«‹ä¸€å€‹ PIL Image å°è±¡
image_path = 'cookie.png'
image = PIL.Image.open(image_path)

# æç¤ºæ–‡æœ¬ï¼Œèªªæ˜ä½ å¸Œæœ›æ¨¡å‹é€²è¡Œçš„æ“ä½œ
prompt = "é€™äº›çœ‹èµ·ä¾†æ˜¯è²·ä¾†çš„é‚„æ˜¯è‡ªè£½çš„ï¼Ÿ"

# ä½¿ç”¨å¤šæ¨¡æ…‹è¼¸å…¥ç”Ÿæˆå…§å®¹
response = model.generate_content(
    # åŒ…å«æ–‡æœ¬å’Œåœ–ç‰‡çš„å…§å®¹
    contents=[prompt, image],
    # å•Ÿç”¨æµå¼è¼¸å‡ºï¼Œå…è¨±é€æ­¥ç”Ÿæˆå›æ‡‰
    stream=True
)

# ç¢ºä¿æ‰€æœ‰æµå¼å›æ‡‰å¡Šéƒ½å·²è™•ç†å®Œç•¢
full_text = ""
for chunk in response:
    # ç´¯ç©ç”Ÿæˆçš„æ–‡æœ¬å¡Š
    full_text += chunk.text

# è¼¸å‡ºæœ€çµ‚ç”Ÿæˆçš„æ–‡æœ¬
print(full_text)
```

è¨­å®šè¿”å› JSON æ ¼å¼å›æ‡‰

```python
# ç”¨æ–¼è§£æ JSON
import json

# è¨­å®šå›æ‡‰ç‚º JSON æ ¼å¼
model = genai.GenerativeModel(
    'gemini-1.5-flash',
    generation_config={
        "response_mime_type": "application/json"
    }
)
# å®šç¾©æç¤ºæ–‡æœ¬ï¼ŒæŒ‡å®šè¿”å› JSON çµæ§‹
prompt = """
    åˆ—å‡º5å€‹ç†±é–€çš„é¤…ä¹¾é£Ÿè­œã€‚
    ä½¿ç”¨é€™å€‹ JSON çµæ§‹ï¼š
        Recipe = {"recipe_name": str}
    return `list[Recipe]`
"""
# ä½¿ç”¨æ¨¡å‹ç”Ÿæˆå…§å®¹
response = model.generate_content(prompt)
# è¼¸å‡ºç”Ÿæˆçš„ JSON æ–‡æœ¬
print(response.text)
# è§£æç”Ÿæˆçš„ JSON æ–‡æœ¬
try:
    # å°‡ JSON æ–‡æœ¬è½‰æ›ç‚º Python å°è±¡
    json_data = json.loads(response.text)
    print("\næ ¼å¼åŒ–å¾Œçš„ JSON æ•¸æ“š:")
    # æ ¼å¼åŒ–ä¸¦è¼¸å‡º JSON æ•¸æ“š
    print(json.dumps(
        json_data,
        indent=4,
        ensure_ascii=False
    ))
except json.JSONDecodeError as e:
    print(f"ç„¡æ³•è§£æ JSON: {e}")
```

ä½¿ç”¨å—æ§ç”Ÿæˆé€²è¡Œ JSON è¿”å›

```python
# å°å…¥ typing_extensions ç”¨æ–¼å®šç¾©å‹åˆ¥
import typing_extensions as typing

# å®šç¾© JSON çµæ§‹
class Recipe(typing.TypedDict):
    # å®šç¾©é£Ÿè­œåç¨±æ¬„ä½
    recipe_name: str

# åˆå§‹åŒ–é«˜ç´šæ¨¡å‹
model = genai.GenerativeModel(
    model_name="models/gemini-1.5-pro"
)

result = model.generate_content(
    # æç¤ºæ–‡æœ¬
    "ä»¥ç¹é«”ä¸­æ–‡åˆ—å‡º5å€‹ç†±é–€çš„é¤…ä¹¾é£Ÿè­œ",  
    # è¨­å®šè¿”å›çš„ JSON çµæ§‹
    generation_config=genai.GenerationConfig(
        response_mime_type="application/json",
        response_schema=list[Recipe]
    )
)
# è¼¸å‡ºç”Ÿæˆçš„ JSON æ–‡æœ¬
print(result.text)

# è§£æç”Ÿæˆçš„ JSON æ–‡æœ¬
try:
    # å°‡ JSON æ–‡æœ¬è½‰æ›ç‚º Python å°è±¡
    json_data = json.loads(result.text)
    print("\næ ¼å¼åŒ–å¾Œçš„ JSON æ•¸æ“š:")
    # æ ¼å¼åŒ–ä¸¦è¼¸å‡º JSON æ•¸æ“š
    print(json.dumps(
        json_data,
        indent=4,
        ensure_ascii=False
    ))
except json.JSONDecodeError as e:
    print(f"ç„¡æ³•è§£æ JSON: {e}")
```
