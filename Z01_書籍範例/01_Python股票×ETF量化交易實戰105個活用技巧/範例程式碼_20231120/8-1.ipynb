"cp950" 是適用於繁體中文的編碼，但如果您開啟檔案的工具（如某些版本的 Microsoft Excel）不支援這種編碼，那麼中文字元可能會顯示為亂碼。

為了解決這個問題，您可以嘗試使用 "utf-8" 編碼來儲存 CSV 檔案。

```python
# 載入必要的套件
import requests
import os
import datetime
import random
import time
import pandas as pd

# 切換工作路徑
os.chdir("./")

# 初始日期設定為目前日期
c_datetime = datetime.datetime.now()

first_write = True

_LAST_DATE = "20230101"

while True:
    c_day = c_datetime.strftime("%Y%m%d")

    # 如果日期小於或等於 _LAST_DATE，則停止循環
    if c_day <= _LAST_DATE:
        break
    print(c_day)
    # 1702746157462 = 2023 年 12 月 16 日 17:02:37 UTC
    url = (
        "https://www.twse.com.tw/rwd/zh/fund/T86?response=json&date="
        + c_day
        + "&selectType=ALL&_=1702746157462"
    )

    response = requests.get(url)

    if response.status_code != 200:
        print(f"Error {response.status_code} while fetching data for date {c_day}")
        c_datetime -= datetime.timedelta(days=1)
        continue

    try:
        jsdata = response.json()
    except ValueError:
        print(f"Failed to decode JSON for the date {c_day}")
        c_datetime -= datetime.timedelta(days=1)
        continue

    if "data" not in jsdata.keys():
        c_datetime -= datetime.timedelta(days=1)
        continue

    data = pd.DataFrame(jsdata["data"], columns=jsdata["fields"])
    data.loc[:, "日期"] = c_day

    if first_write:
        data.to_csv(
            "三大法人爬蟲資料.csv", 
            encoding="utf-8",
            index=False
        )
        first_write = False
    else:
        data.to_csv(
            "三大法人爬蟲資料.csv",
            encoding="utf-8",
            mode="a",
            index=False,
            header=False
        )

    # 隨機休息 1-5 秒
    time.sleep(random.randint(1, 5))

    # 遞減一天
    c_datetime -= datetime.timedelta(days=1)
    
```
