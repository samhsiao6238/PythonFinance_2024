FinMind 整合 ffn

ffn 的 index 是日期，但是 FinMind 的 index 是流水號，date 是其中一個欄位而且是 string 格式，所以我們要將 FinMind 的 index 改成日期格式的 date

先確認有資料

```python
stock_data
```

以下這樣會報錯，因為在 Pandas 中，一旦某個列被設定為索引，它就不再被視為普通的欄位。 這表示如果您嘗試使用 set_index 方法將 date 欄位設為索引，將會導致 KeyError，因為 date 已不再是一個可識別的資料列，而是索引。

```python
stock_data.set_index("date", inplace=True)
```

轉回索引

```python
stock_data.reset_index(inplace=True)
stock_data
```

```python
import pandas as pd
stock_data = stock_data.set_index(pd.DatetimeIndex(pd.to_datetime(stock_data.index)))
stock_data
```
