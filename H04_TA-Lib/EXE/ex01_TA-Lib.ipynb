模擬 180 天的股市資料

```python
import pandas as pd
import numpy as np
import openpyxl

# 設定隨機數種子以保持數據的一致性
np.random.seed(0)

# 模擬 180 天的股市資料
dates = pd.date_range(start="2023-01-01", periods=180, freq='B')  # 僅包含工作日
open_prices = np.random.uniform(low=100, high=200, size=len(dates))
high_prices = open_prices + np.random.uniform(low=0, high=10, size=len(dates))
low_prices = open_prices - np.random.uniform(low=0, high=10, size=len(dates))
close_prices = np.random.uniform(low=low_prices, high=high_prices)

# 建立 DataFrame
stock_data = pd.DataFrame({
    'Date': dates,
    'Open': open_prices,
    'High': high_prices,
    'Low': low_prices,
    'Close': close_prices
})

# 將日期設置為索引
stock_data.set_index('Date', inplace=True)

# 儲存數據到 Excel 文件
excel_file_path = 'stock_data.xlsx'  # 您可以更改路徑
stock_data.to_excel(excel_file_path)

print(f"股市資料已儲存到 {excel_file_path}")
```

close_prices 是一個包含收盤價的 NumPy 數組

```python
# 首先讀取剛才儲存的 Excel 文件
stock_data_read = pd.read_excel('stock_data.xlsx', index_col=0)

# 從讀取的數據中提取收盤價
close_prices = np.array(stock_data_read['Close'])

# 展示一部分收盤價數據
# close_prices[:10]  # 顯示前 10 個收盤價
```

```python
# 計算 30 日簡單移動平均線
moving_average = talib.SMA(close_prices, timeperiod=30)
moving_average
```

指數移動平均（EMA）

```python
EMA = talib.EMA(close_prices, timeperiod=30)
EMA
```

相對強弱指數（RSI）

```python
RSI = talib.RSI(close_prices, timeperiod=14)
RSI
```

隨機指標（Stochastic Oscillator）

```python
slowk, slowd = talib.STOCH(high, low, close, fastk_period=5, slowk_period=3, slowk_matype=0, slowd_period=3, slowd_matype=0)
```

平均方向性指數（ADX）

```python
ADX = talib.ADX(high, low, close, timeperiod=14)
```
