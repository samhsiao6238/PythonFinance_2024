## 運動相簿照片基本下載

### 載入相關的模組

```python
import requests,os
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

### 設定相簿網址

```python
# 第3屆埔里跑 Puli Power 山城派對馬拉松  向善橋(約34K)
url = 'http://tw.running.biji.co/index.php?q=album&act=photo_list\
&album_id=30668&cid=5791&type=album\
&subtitle=第3屆埔里跑 Puli Power 山城派對馬拉松-向善橋(約34K)'
```

### 取得相簿名稱

```python
html = requests.get(url)
soup=BeautifulSoup(html.text,'html.parser')  
title = soup.find('h1',{'class':'album-title flex-1'}).text.strip()
```

### 以data參數提出post請求

```python
url = 'https://running.biji.co/index.php?pop=ajax\
&func=album&fename=load_more_photos_in_listing_computer'

payload = {'type': 'place', 'rows': '0','need_rows': '20', 
           'cid': '5791','album_id': '30668'}
html = requests.post(url, data=payload)
# 在回應頁面中找出所有照片連結
soup = BeautifulSoup(html.text, 'html.parser')
```

### 以標題建立目錄儲存圖片

```python
images_dir=title + "/"
if not os.path.exists(images_dir):
    os.mkdir(images_dir)
```

### 下載圖片

```python
# 處理所有 <img> 標籤    
photos = soup.select('.photo_img') 
n=0 # 記錄下載圖片數量
for photo in photos:
    # 讀取 src 屬性內容
    src=photo['src']
    # 讀取 .jpg 檔
    if src != None and ('.jpg' in src):
        # 設定圖檔完整路徑
        full_path = src            
        filename = full_path.split('/')[-1]  # 取得圖檔名
        print(full_path)
        # 儲存圖片
        try:
            image = urlopen(full_path)
            with open(os.path.join(images_dir,filename),'wb') as f:
                f.write(image.read())  
            n+=1
        except:
            print("{} 無法讀取!".format(filename))
            
print("共下載",n,"張圖片") 
```

## 運動相簿照片批次下載

### 批次下載程式碼

```python
import requests,os
from bs4 import BeautifulSoup
from urllib.request import urlopen

# 第3屆埔里跑 Puli Power 山城派對馬拉松  向善橋(約34K)
url = 'http://tw.running.biji.co/index.php?q=album&act=photo_list\
&album_id=30668&cid=5791&type=album\
&subtitle=第3屆埔里跑 Puli Power 山城派對馬拉松-向善橋(約34K)'
html = requests.get(url)

soup=BeautifulSoup(html.text,'html.parser')  
title = soup.find('h1',{'class':'album-title flex-1'}).text.strip() + "_全部相片"

url = 'https://running.biji.co/index.php?pop=ajax\
&func=album&fename=load_more_photos_in_listing_computer'

# 在回應頁面中找出所有照片連結
soup = BeautifulSoup(html.text, 'html.parser')

# 以標題建立目錄儲存圖片
images_dir=title + "/"
if not os.path.exists(images_dir):
    os.mkdir(images_dir)

n=0 # 記錄下載圖片數量
for i in range(200):
    payload = {'type': 'place', 'rows': str(i*20), 'need_rows': '20',
               'cid': '5791','album_id': '30668'}
    html = requests.post(url, data=payload)
    soup = BeautifulSoup(html.text, 'html.parser')
    # 處理所有 <img> 標籤 
    photos = soup.select('.photo_img')
    for photo in photos:
        # 讀取 src 屬性內容
        src=photo['src']
        # 讀取 .jpg 檔
        if src != None and ('.jpg' in src):
            # 設定圖檔完整路徑
            full_path = src            
            filename = full_path.split('/')[-1]  # 取得圖檔名
            # 儲存圖片
            try:
                image = urlopen(full_path)
                with open(os.path.join(images_dir,filename),'wb') as f:
                    f.write(image.read())  
                n+=1
                if n % 50 ==0:
                  print("n=",n)
                if n>=1000: # 最多下載 1000 張
                    break   # 離開內部 for 迴圈
            except:
                print("{} 無法讀取!".format(filename))  
    if n>=1000: # 最多下載 1000 張
        break   # 離開外部 for 迴圈      
                
print("共下載",n,"張圖片") 
```

## 非同步模組- concurrent.futures

```python
from concurrent.futures import ThreadPoolExecutor
 
def show(fruit):
    print(fruit)

fruits = ('西瓜', '百香果', '香蕉', '橘子', '蘋果')
with ThreadPoolExecutor(max_workers=5) as executor:
    results = executor.map(show, fruits)
```

### 運動相簿照片非同步下載

```python
import requests,os
from bs4 import BeautifulSoup
from urllib.request import urlopen
import concurrent.futures
import time

def scraper(row): 
    global n    
    payload = {'type': 'place', 'rows': str(row*20), 'need_rows': '20',
               'cid': '5791','album_id': '30668'}   
    html = requests.post(url, data=payload)
    # 在回應頁面中找出所有照片連結
    soup = BeautifulSoup(html.text, 'html.parser')
    
    # 處理所有 <img> 標籤    
    photos = soup.select('.photo_img') 
    for photo in photos:
        # 讀取 src 屬性內容
        src=photo['src']
        # 讀取 .jpg 檔
        if src != None and ('.jpg' in src):
            # 設定圖檔完整路徑
            full_path = src            
            filename = full_path.split('/')[-1]  # 取得圖檔名
            # print(full_path)
            # 儲存圖片
            try:
                image = urlopen(full_path)
                with open(os.path.join(images_dir,filename),'wb') as f:
                    f.write(image.read())  
                n+=1
                if n % 50 ==0:
                  print("n=",n)
            except:
                print("{} 無法讀取!".format(filename))   

start_time = time.time()  # 開始時間
url = 'http://tw.running.biji.co/index.php?q=album&act=photo_list\
&album_id=30668&cid=5791&type=album\
&subtitle=第3屆埔里跑 Puli Power 山城派對馬拉松-向善橋(約34K)'
html = requests.get(url)

soup=BeautifulSoup(html.text,'html.parser')  
title = soup.find('h1',{'class':'album-title flex-1'}).text.strip() + "_全部相片(多執行緒)"
# 以標題建立目錄儲存圖片
images_dir=title + "/"
if not os.path.exists(images_dir):
    os.mkdir(images_dir) 
                
n=0 # 記錄下載圖片數量
# 第3屆埔里跑 Puli Power 山城派對馬拉松  向善橋(約34K)   
url = 'https://running.biji.co/index.php?pop=ajax\
&func=album&fename=load_more_photos_in_listing_computer'

# 同時建立及啟用100個執行緒
rows=list(range(0, 50))
with concurrent.futures.ThreadPoolExecutor(max_workers=100) as executor:
    executor.map(scraper,rows)  
          
print("共下載",n,"張圖片") 

end_time = time.time() # 結束時間
print(f"使用時間：{end_time - start_time} 秒")
```
