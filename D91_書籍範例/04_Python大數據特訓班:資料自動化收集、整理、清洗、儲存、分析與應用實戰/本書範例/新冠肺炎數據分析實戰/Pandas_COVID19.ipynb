# Pandas資料讀取顯示及應用

【以嚴重特殊傳染性肺炎(COVID-19)資料為例】

* 資料網站：https://nidss.cdc.gov.tw/ch/NIDSS_DiseaseMap.aspx?dc=1&dt=5&disease=19CoV
* CSV來源：http://od.cdc.gov.tw/eic/Weekly_Age_County_Gender_19CoV.csv

```python
import pandas as pd
df = pd.read_csv('https://od.cdc.gov.tw/eic/Weekly_Age_County_Gender_19CoV.csv')
```

# Pandas 資料排序

```python
df.sort_values(['診斷年份','診斷週別'], ascending=[True, False])
```

```python
df1 = df.sort_values(['診斷年份','診斷週別'], ascending=[True, False])
```

```python
df.sort_values(['診斷年份','診斷週別'], ascending=[True, False], inplace=True)
```

```python
df
```

# Pandas 資料篩選

```python
df['縣市'] == "台北市"
```

```python
df.縣市 == "台北市"
```

```python
df[df['縣市'] == "台北市"]
```

```python
(df['縣市'] == "台北市") & (df["性別"] == "M")
```

```python
df[(df['縣市'] == "台北市") & (df["性別"] == "M")]
```

```python
df[(df['縣市'] == "台北市") | (df['縣市'] == "台中市") | (df['縣市'] == "高雄市")]
```

```python
df[df['縣市'].isin(["台北市","台中市","高雄市"])]
```

# Pandas 群組統計(group by)

```python
df.groupby('縣市')['確定病例數'].sum()
```

```python
df.groupby('年齡層')['確定病例數'].sum()
```

```python
df.groupby('縣市')['確定病例數'].count()
```

```python
df.groupby('縣市')['確定病例數'].min()
```

```python
df.groupby('縣市')['確定病例數'].agg(['sum','count','max','min','mean'])
```

# Pandas 繪圖

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'mingliu'
df.groupby('縣市')['確定病例數'].sum().plot(kind='bar')
```

```python
df.groupby('年齡層')['確定病例數'].sum().plot(kind='pie')
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'mingliu'
df.groupby('診斷週別')['確定病例數'].sum().plot(kind='line', linestyle="--", marker="s", grid=True)
```
