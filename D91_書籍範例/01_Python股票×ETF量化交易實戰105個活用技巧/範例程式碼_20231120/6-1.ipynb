```python
# 載入必要套件
from Data import getDataFM
from BackTest import ChartTrade
import mplfinance as mpf
from talib.abstract import SMA, WMA, EMA

# 取得回測資料
prod = "0050"
data = getDataFM(prod, "2017-01-01", "2023-12-15")

# 計算指數移動平均線
data["ema"] = EMA(data['close'], timeperiod=120)
data["wma"] = WMA(data['close'], timeperiod=120)
data["sma"] = SMA(data['close'], timeperiod=120)

# 繪製副圖
addp = []
addp.append(mpf.make_addplot(data["ema"], color='blue'))
addp.append(mpf.make_addplot(data["wma"], color='green'))
addp.append(mpf.make_addplot(data["sma"], color='red'))

# 繪製K線圖與交易明細
ChartTrade(data, addp=addp)
```
