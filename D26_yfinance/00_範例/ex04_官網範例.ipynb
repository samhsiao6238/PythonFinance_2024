官網的範例

```python
# 導入 yfinance 庫
import yfinance as yf  
# 建立一個 Ticker 物件代表微軟公司 (股票代碼：MSFT)
msft = yf.Ticker("MSFT")  

# 獲取所有股票信息
# 獲取微軟公司的所有股票信息
msft.info  

# 獲取歷史市場數據
# 獲取微軟公司過去一個月的歷史市場數據
hist = msft.history(period="1mo")  

# 顯示關於歷史數據的元信息（需要先調用 history() 方法）
# 顯示歷史數據的元信息
msft.history_metadata  

# 顯示行動（股息、分股、資本收益）
# 顯示微軟公司的所有行動（包括股息和分股）
msft.actions  
# 顯示微軟公司的股息
msft.dividends  
# 顯示微軟公司的分股
msft.splits  
# 顯示資本收益（僅限於共同基金和 ETFs）
msft.capital_gains  

# 顯示股票數量
# 從 2022 年 1 月 1 日開始的股票數量
msft.get_shares_full(start="2022-01-01", end=None)  

# 顯示財務報表：
# - 收益表
# 年度收益表
msft.income_stmt  
# 季度收益表
msft.quarterly_income_stmt  
# - 資產負債表
# 年度資產負債表
msft.balance_sheet  
# 季度資產負債表
msft.quarterly_balance_sheet  
# - 現金流量表
# 年度現金流量表
msft.cashflow  
# 季度現金流量表
msft.quarterly_cashflow  
# 更多選項見 `Ticker.get_income_stmt()`

# 顯示持有者
# 主要持有者
msft.major_holders  
# 機構持有者
msft.institutional_holders  
# 共同基金持有者
msft.mutualfund_holders  

# 顯示未來和歷史盈利日期，預設最多顯示接下來 4 個季度和最近 8 個季度的數據。
# 注意：如果需要更多數據，使用 msft.get_earnings_dates(limit=XX) 並增加 limit 參數。
msft.earnings_dates

# 顯示 ISIN 代碼 - *實驗性功能*
# ISIN = 國際證券識別碼
print(msft.isin)

# 顯示期權到期日
# 顯示微軟公司的期權到期日
print(msft.options)

# 顯示新聞
# 顯示有關微軟公司的最新新聞
print(msft.news)

# 獲取特定到期日的期權鏈
# 將 'YYYY-MM-DD' 替換為 '2023-12-15'
opt = msft.option_chain('2023-12-15')  
# 數據可通過：opt.calls, opt.puts 獲取
opt
```

將 requests_cache 與速率限制相結合，以避免觸發雅虎的速率限制器/阻止程序，從而損壞資料。

```python
%pip install requests_ratelimiter
```

```python
from requests import Session
from requests_cache import CacheMixin, SQLiteCache
from requests_ratelimiter import LimiterMixin, MemoryQueueBucket
from pyrate_limiter import Duration, RequestRate, Limiter
class CachedLimiterSession(CacheMixin, LimiterMixin, Session):
    pass

session = CachedLimiterSession(
    limiter=Limiter(RequestRate(2, Duration.SECOND*5)),  # max 2 requests per 5 seconds
    bucket_class=MemoryQueueBucket,
    backend=SQLiteCache("yfinance.cache"),
)
```
