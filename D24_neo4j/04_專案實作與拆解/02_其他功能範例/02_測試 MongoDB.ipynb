使用 tls=True 和 tlsAllowInvalidCertificates=True 參數可以避免 SSL 驗證問題，這是因為這些參數告訴 MongoClient 在建立 SSL/TLS 連接時忽略證書驗證錯誤。

```python
from pymongo.mongo_client import MongoClient

uri = "mongodb+srv://sam6238:sam112233@cluster0.yhwvqqt.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"

# 建立連線
client = MongoClient(
    uri,
    tls=True,
    tlsAllowInvalidCertificates=True
)

# 檢查連線
try:
    client.admin.command('ping')
    print("連線成功。")
except Exception as e:
    print(e)
```

```python
from pymongo import MongoClient
import certifi

# MongoDB 连接配置
URL = "mongodb+srv://sam6238:sam112233@cluster0.yhwvqqt.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"
database_name = "MyDatabase2024"
collection_name = "MyCollection2024"

# 连接到 MongoDB
client = MongoClient(URL, tlsCAFile=certifi.where(), tls=True, tlsAllowInvalidCertificates=True)
db = client[database_name]
collection = db[collection_name]

# 获取集合中的一个样本文档
sample_doc = collection.find_one()

if sample_doc:
    print("集合中的字段及其类型如下：")
    for field, value in sample_doc.items():
        print(f"{field}: {type(value).__name__}")
else:
    print("集合为空或不存在。")
```
