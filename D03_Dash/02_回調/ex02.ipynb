取得資料


```python
import pandas as pd
import requests

url = "https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json"

# 發送請求並獲取數據
response = requests.get(url)

# 檢查請求是否成功
if response.status_code == 200:
    # 將資料解析為JSON
    data = response.json()
    # 轉換為DataFrame
    df = pd.DataFrame(data)
    # 列印前幾行數據
    print(df.head())
else:
    print("無法下載數據，狀態碼：", response.status_code)
```

建立映射

```python
column_mapping = {
    "sno": "站點編號",
    "sna": "站點名稱",
    "tot": "車位總數",
    "sbi": "可借車數",
    "sarea": "所在區域",
    "mday": "資料更新時間",
    "lat": "緯度",
    "lng": "經度",
    "ar": "地址",
    "sareaen": "區域英文名稱",
    "snaen": "站點英文名稱",
    "aren": "地址英文",
    "bemp": "空位數量",
    "act": "啟用狀態",
    "srcUpdateTime": "來源更新時間",
    "updateTime": "更新時間",
    "infoTime": "訊息時間",
    "infoDate": "訊息日期"
}
```

```python
# 應用映射以更改欄位名稱
df.rename(columns=column_mapping, inplace=True)
df
```

```python
# 假設 df 是先前處理過的 DataFrame
# 選擇所需的列
selected_columns = ["站點名稱", "車位總數", "可借車數", "所在區域"]

# 使用 selected_columns 選擇 DataFrame 的子集
selected_df = df[selected_columns]
# 顯示選擇列的前幾行數據
selected_df.head()
```

```python
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd
```

```python
# 建立 Dash 應用
app = dash.Dash(__name__)

# 定義佈局
app.layout = html.Div(
    [
        dcc.Dropdown(
            id="area-dropdown",
            options=[{"label": area, "value": area} for area in selected_df["所在區域"].unique()],
            value=selected_df["所在區域"].unique()[0],  # 預設值
        ),
        dcc.Graph(id="youbike-graph"),
    ]
)

# 定義callback來更新圖表
@app.callback(Output("youbike-graph", "figure"), Input("area-dropdown", "value"))
def update_graph(selected_area):
    # 根據選取的區域過濾數據
    filtered_df = selected_df[selected_df["所在區域"] == selected_area]

    # 產生圖表
    fig = px.bar(filtered_df, x="站點名稱", y=["車位總數", "可借車數"], barmode="group", title="YouBike站點數據")

    return fig

# 運行應用
if __name__ == "__main__":
    app.run_server(debug=True)
```
